```
#import random module
import random
word_list = ['boolean', 'loop', 'float', 'tuple', 'dictionary', 'function', 'string', 'mutable', 'module', 'operator', 'integer', 'variable']

#need to define function to pull word from the list
def pull_word():
  word = random.choice(word_list)
  word = word.lower() #lower is used to simplify code logic (no need to account for differences in typing)
  return word 

#function if input has a number it is invalid
def hasNumbers(inputString):
    return any(char.isdigit() for char in inputString)

#create hangman base/gallow
def display(attempts): #print gallow based on number of attempts
  if (attempts == 1):
    print(" _____\n |    |\n |    O\n |    \n |    \n |\n_|______")
  elif (attempts == 2):
 	  print(" _____\n |    |\n |    O \n |    |\n |    |\n |\n_|______")
  elif (attempts == 3):
	  print(" _____\n |    |\n |    O\n |   \|\n |    |\n |\n_|______")
  elif (attempts == 4):
    print(" _____\n |    |\n |    O\n |   \|/\n |    |\n |\n_|______")
  elif (attempts == 5):
    print(" _____\n |    |\n |    O\n |   \|/\n |    |\n |   / \n_|______")
  elif (attempts == 6):
   print(" _____\n |    |\n |    O\n |   \|/\n |    |\n |   / \ \n_|______")
  else:
    print(" _____\n |    |\n |\n |\n |\n |\n_|______")

```

```

def game_play(word):

  attempts = 0
  display(attempts) #links gallow with game play
  word_list = list(word)
  guess_this = '_' * len(word) #number of spaces
  guess_this_list = list(guess_this) #spaces to be filled
  letters_guessed = list(guess_this) #correct letters guessed
  guesses = [] #guess list

  print("" + ' '.join(guess_this_list))

  while attempts < 6: 

      guess = input("Guess a letter: ") #input to guess a letter
      guess = guess.lower()  #turn into lower case 
  
      if len(guess) > 1:                                 #check if user entered more than one letter
          print ("Enter one letter at a time.")
      elif guess in guesses:                             #check if the letter has already been guessed
          print ("You have already guessed this letter. Guessed letters are:")
          print(' '.join(guesses))
      elif guess == "":                                  #if user just pressed space
          print ("Please enter a letter. ") 
      elif hasNumbers(guess):                            #if user enters number
          print("Enter letter only. ")
      elif not guess.isalpha():
          print("Enter letter only. ")                   #enter character only!
      else:                                              #check if letter in word                              
          guesses.append(guess)                          #adds new guess to guesses     
          char = 0
          while char < len(word):                
              if guess == word[char]:    
                  letters_guessed[char] = word_list[char]
              char = char + 1                            #put letter in slot, added right letter to letters_guessed

          if letters_guessed == guess_this_list:
              print("Your guess is wrong!")              
              attempts = attempts + 1
              display(attempts)                           #showing the hangman

              if attempts < 6:
                  print("Try again.")
                  print("Guesses left: " + str(6-attempts)) 
                  print(' '.join(guess_this_list))        #print attempts left if < 6

              if attempts == 6:                           #alotted 6 attempts, this is game over
                  print("Game over! You lost. The word is " + word)
          
          elif word_list != guess_this_list:              #if the letters in the word do not equal the slots list (haven't guessed word yet)
              guess_this_list = letters_guessed[:]        #reset the letters_guessed list to be the same
              print(' '.join(guess_this_list))       

              if word_list == guess_this_list:
                  print("You win!\nCongratulations.")
                  break
              else:
                  print("Good guess!")



```
